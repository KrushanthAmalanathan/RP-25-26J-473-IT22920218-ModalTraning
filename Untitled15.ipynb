```python
from pathlib import Path
import random, shutil, yaml, os
```

```python
SOURCE_ROOT = Path("/content/drive/MyDrive/Vehicle_classification/vehicle classification.v3i.yolov11")

# We'll build a separate subset folder (SAFE)
SUBSET_ROOT = Path("/content/drive/MyDrive/Vehicle_classification/vehicle_subset_3000")

# Original folders (we will read from these)
SRC_IMG_DIR = SOURCE_ROOT / "train/images"
SRC_LBL_DIR = SOURCE_ROOT / "train/labels"
```

```python
NUM_IMAGES = 3000
SEED = 42
TRAIN_RATIO = 0.80
VALID_RATIO = 0.10
TEST_RATIO  = 0.10   # must make total = 1.0

# If subset already exists, you can set REBUILD_SUBSET=False to reuse it
REBUILD_SUBSET = True

img_exts = {".jpg", ".jpeg", ".png", ".webp", ".bmp"}
```

```python
def list_images(folder: Path):
    return [p for p in folder.iterdir() if p.is_file() and p.suffix.lower() in img_exts]

def ensure_folders(root: Path):
    for split in ["train", "valid", "test"]:
        (root / split / "images").mkdir(parents=True, exist_ok=True)
        (root / split / "labels").mkdir(parents=True, exist_ok=True)

def clear_folder(folder: Path):
    if folder.exists():
        for f in folder.rglob("*"):
            if f.is_file():
                f.unlink()

def copy_pair(img_path: Path, dst_split: str):
    """Copy image + corresponding label into subset split folder."""
    lbl_path = SRC_LBL_DIR / f"{img_path.stem}.txt"
    if not lbl_path.exists():
        return False

    shutil.copy2(img_path, SUBSET_ROOT / dst_split / "images" / img_path.name)
    shutil.copy2(lbl_path, SUBSET_ROOT / dst_split / "labels" / lbl_path.name)
    return True

def count_imgs(folder: Path):
    return len([p for p in folder.iterdir() if p.is_file() and p.suffix.lower() in img_exts])
```

```python
orig_yaml_path = SOURCE_ROOT / "data.yaml"
if not orig_yaml_path.exists():
    raise FileNotFoundError(f"âŒ Original data.yaml not found at: {orig_yaml_path}")

orig_data = yaml.safe_load(open(orig_yaml_path))

# Robust: support both 'names' list and dict
orig_names = orig_data.get("names")
if isinstance(orig_names, dict):
    # convert dict {0:"a",1:"b"} to list
    names_list = [orig_names[i] for i in range(len(orig_names))]
else:
    names_list = orig_names

nc = int(orig_data.get("nc", len(names_list)))

print("âœ… Original classes:", nc)
print("âœ… Class names:", names_list)
```

```python
ensure_folders(SUBSET_ROOT)

# Guard file so we know subset is already prepared
SUBSET_GUARD = SUBSET_ROOT / ".subset_done"

if REBUILD_SUBSET:
    # Clean existing subset files (safe; only deletes subset copies)
    clear_folder(SUBSET_ROOT / "train")
    clear_folder(SUBSET_ROOT / "valid")
    clear_folder(SUBSET_ROOT / "test")
    if SUBSET_GUARD.exists():
        SUBSET_GUARD.unlink()

if not SUBSET_GUARD.exists():
    all_imgs = list_images(SRC_IMG_DIR)
    if len(all_imgs) < NUM_IMAGES:
        raise ValueError(f"âŒ Not enough images in source train/images. Found {len(all_imgs)}, need {NUM_IMAGES}")

    random.seed(SEED)
    all_imgs = sorted(all_imgs, key=lambda x: x.name)  # deterministic
    random.shuffle(all_imgs)

    selected = all_imgs[:NUM_IMAGES]

    n_train = int(NUM_IMAGES * TRAIN_RATIO)
    n_valid = int(NUM_IMAGES * VALID_RATIO)
    # rest -> test
    train_imgs = selected[:n_train]
    valid_imgs = selected[n_train:n_train + n_valid]
    test_imgs  = selected[n_train + n_valid:]

    moved_train = moved_valid = moved_test = 0
    skipped = 0

    for img in train_imgs:
        ok = copy_pair(img, "train")
        moved_train += int(ok)
        skipped += int(not ok)

    for img in valid_imgs:
        ok = copy_pair(img, "valid")
        moved_valid += int(ok)
        skipped += int(not ok)

    for img in test_imgs:
        ok = copy_pair(img, "test")
        moved_test += int(ok)
        skipped += int(not ok)

    SUBSET_GUARD.write_text(
        f"subset={NUM_IMAGES}\nseed={SEED}\ntrain={moved_train}\nvalid={moved_valid}\ntest={moved_test}\nskipped={skipped}\n"
    )

    print("âœ… Subset created")
    print("Train copied:", moved_train)
    print("Valid copied:", moved_valid)
    print("Test  copied:", moved_test)
    print("Skipped (missing label):", skipped)
else:
    print("â„¹ï¸ Subset already exists. Reusing it.")

# Verify subset counts
print("\nğŸ“Œ SUBSET counts:")
print("Train:", count_imgs(SUBSET_ROOT / "train/images"))
print("Valid:", count_imgs(SUBSET_ROOT / "valid/images"))
print("Test :", count_imgs(SUBSET_ROOT / "test/images"))
```

```python
subset_yaml = {
    "path": str(SUBSET_ROOT),
    "train": "train/images",
    "val": "valid/images",
    "test": "test/images",
    "nc": nc,
    "names": names_list
}

subset_yaml_path = SUBSET_ROOT / "data.yaml"
with open(subset_yaml_path, "w") as f:
    yaml.dump(subset_yaml, f, sort_keys=False)

print("\nâœ… Subset data.yaml saved at:", subset_yaml_path)
print(open(subset_yaml_path).read())
```

```python
from ultralytics import YOLO

# Auto GPU/CPU
device = 0
try:
    import torch
    if not torch.cuda.is_available():
        device = "cpu"
except:
    device = "cpu"
print("âœ… Using device:", device)

model = YOLO("yolo11n.pt")  # try yolo11s.pt later for better accuracy

model.train(
    data=str(subset_yaml_path),
    epochs=40,
    imgsz=640,
    batch=16,       # if CUDA OOM -> 8 or 4
    device=device,
    workers=0,      # âœ… Drive stability
    cache=False,    # âœ… avoid cache errors on Drive
    project="yolo11_runs",
    name="vehicle_subset_3000",
    patience=15
)
```

```python
best_path = "yolo11_runs/vehicle_subset_3000/weights/best.pt"
best = YOLO(best_path)

print("\n===== VALIDATION (valid) =====")
best.val(data=str(subset_yaml_path), split="val", workers=0)

print("\n===== TEST (test) =====")
best.val(data=str(subset_yaml_path), split="test", workers=0)
```

```python
print("\n===== PREDICT ON TEST (saved) =====")
best.predict(
    source=str(SUBSET_ROOT / "test/images"),
    conf=0.25,
    save=True,
    project="yolo11_runs",
    name="subset_3000_test_predictions"
)
```

```python
save_to = SUBSET_ROOT / "best.pt"
shutil.copy(best_path, str(save_to))
print("\nâœ… best.pt saved to:", save_to)

# Optional export
# best.export(format="onnx")
```
